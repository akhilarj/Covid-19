### Covid 19 Numbers Extraction, Cleaning and EDA

```python
# imports
import os
import re
import pandas as pd
import numpy as np
import datetime as dt 
import matplotlib.pyplot as plt
import seaborn as sns

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
## read csv dataset
covid19 = pd.read_csv('../datasets/covid19.csv')
covid19.head()
```

```python
covid19.info()
```

```python
# rename columns
# covid19.columns = covid19.columns.map(lambda x: x.replace(' ', '_'))
# covid19.columns = covid19.columns.map(lambda x: x.replace('/', '_'))
covid19.columns = covid19.columns.map(lambda x: x.lower())
covid19.info()
```

```python
# convert date fields to correct dtypes
# covid19.observationdate = pd.to_datetime(covid19.observationdate)
covid19.last_update = pd.to_datetime(covid19.last_update)
covid19.info()
```

```python
# dropping non-relevant columns
covid19.drop(['fips','admin2'], axis = 1, inplace = True)
covid19.info()
```

```python
covid19.confirmed.sum(), covid19.deaths.sum(), covid19.recovered.sum()
```

```python
#no of cases reported each day
df_by_date=covid19.groupby(['last_update']).sum().reset_index(drop=None) 
df_by_date
```

### Visualizations

```python
df_by_date=covid19.groupby(['last_update']).sum().reset_index(drop=None)
df_by_date['daily_cases']=df_by_date.confirmed
df_by_date['daily_deaths']=df_by_date.deaths
df_by_date['daily_recoveries']=df_by_date.recovered
df_by_date
```

```python
#setting the image size
plt.rcParams['figure.figsize']=(15,7)

#plotting the cases by date
df_by_date.groupby(['last_update']).sum().plot(kind='bar')
plt.xticks(rotation=60)
plt.ylabel('Number of cases',fontsize=15)
plt.xlabel('Dates',fontsize=15)
```
