### Modeling

```python
## read csv dataset
covid19 = pd.read_csv('../datasets/covid_19_data.csv')
covid19.head()
```

```python
train_data = pd.read_csv('../datasets/train.csv')
```

#### Pre-processing data before to prapare it for modeling

    1. Fill NA values in Province_State column 
    2. Replace special characters in Country_Region & Province_State columns
    3. Change Date column to datetime data type
    4. Calculate day counts from initial date of data available
    5. Scale data

```python
train_data['Country_Region']= train_data['Country_Region'].str.replace("'", "")
train_data['Province_State']= train_data['Province_State'].str.replace("'", "")
```

```python
train_data["ConfirmedCases"] = train_data["ConfirmedCases"].astype("float")
train_data["Fatalities"] = train_data["Fatalities"].astype("float")
train_data["Country_Region"] = [ row.Country_Region.replace("'","").strip(" ") if row.Province_State=="" else row.Country_Region+"_"+str(row.Province_State).replace("'","").strip(" ") for idx,row in train_data.iterrows()]
```

```python
train_data.info()
train_data.tail()
```

```python
from tqdm import tqdm
train_data = train_data.query("Date > '2020-01-22'and Date < '2020-04-01'")
days_in_sequence = 21

trend_list = []

with tqdm(total=len(list(train_data.Country_Region.unique()))) as pbar:
    for country in train_data.Country_Region.unique():
        for province in train_data.query(f"Country_Region=='{country}'").Province_State.unique():
            province_df = train_data.query(f"Country_Region=='{country}' and Province_State=='{province}'")
            
            for i in range(0,len(province_df),int(days_in_sequence/3)):
                if i+days_in_sequence<=len(province_df):
                    #prepare all the temporal inputs
                    infection_trend = [float(x) for x in province_df[i:i+days_in_sequence-1].ConfirmedCases.values]
                    fatality_trend = [float(x) for x in province_df[i:i+days_in_sequence-1].Fatalities.values]

                    expected_cases = float(province_df.iloc[i+days_in_sequence-1].ConfirmedCases)
                    expected_fatalities = float(province_df.iloc[i+days_in_sequence-1].Fatalities)

                    trend_list.append({"infection_trend":infection_trend,
                                     "fatality_trend":fatality_trend,
                                     "expected_cases":expected_cases,
                                     "expected_fatalities":expected_fatalities})
        pbar.update(1)
trend_data = pd.DataFrame(trend_list)
```

```python
from sklearn.utils import shuffle
trend_data["case_inputs"] = [np.asarray([trends["infection_trend"],trends["fatality_trend"]]) for idx,trends in trend_data.iterrows()]

trend_data = shuffle(trend_data)
```

```python
i=0
temp_df = pd.DataFrame()
for idx,row in trend_data.iterrows():
    if sum(row.infection_trend)>0:
        temp_df = temp_df.append(row)
    else:
        if i<25:
            temp_df = temp_df.append(row)
            i+=1
trend_data = temp_df
```

```python
trend_data.head()
```

```python
sequence_length = 20
training_percentage = 0.9

training_item_count = int(len(trend_data)*training_percentage)
validation_item_count = len(trend_data)-int(len(trend_data)*training_percentage)
training_df = trend_data[:training_item_count]
validation_df = trend_data[training_item_count:]
```

```python
X_cases_train = np.asarray(np.transpose(np.reshape(np.asarray([np.asarray(x) for x in training_df["case_inputs"].values]),(training_item_count,2,sequence_length)),(0,2,1) )).astype(np.float32)
Y_expcases_train = np.asarray([np.asarray(x) for x in training_df["expected_cases"]]).astype(np.float32)
Y_expfatalities_train = np.asarray([np.asarray(x) for x in training_df["expected_fatalities"]]).astype(np.float32)
```

```python
X_cases_test = np.asarray(np.transpose(np.reshape(np.asarray([np.asarray(x) for x in validation_df["case_inputs"]]),(validation_item_count,2,sequence_length)),(0,2,1)) ).astype(np.float32)
Y_expcases_test = np.asarray([np.asarray(x) for x in validation_df["expected_cases"]]).astype(np.float32)
Y_expfatalities_test = np.asarray([np.asarray(x) for x in validation_df["expected_fatalities"]]).astype(np.float32)
```

**Model trials:**
1. Dropout for both cases and fatalities : 0.3
2. Dropout for both cases and fatalities : 0.05

```python
from tensorflow.keras import Input
from tensorflow.keras import layers
from tensorflow.keras.models import Model
#temporal input branch
cases_input_layer = Input(shape=(sequence_length,2))
main_rnn_layer = layers.LSTM(64, return_sequences=True, recurrent_dropout=0.2)(cases_input_layer)

#cases output branch
rnn_c = layers.LSTM(32)(main_rnn_layer)
dense_c = layers.Dense(128)(rnn_c)
dropout_c = layers.Dropout(0.05)(dense_c)
cases = layers.Dense(1, activation='relu',name="cases")(dropout_c)

#fatality output branch
rnn_f = layers.LSTM(32)(main_rnn_layer)
dense_f = layers.Dense(128)(rnn_f)
dropout_f = layers.Dropout(0.05)(dense_f)
fatalities = layers.Dense(1, activation='relu', name="fatalities")(dropout_f)


model = Model([cases_input_layer], [cases,fatalities])

model.summary()
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint
callbacks = [ReduceLROnPlateau(monitor='val_loss', patience=4, verbose=1, factor=0.6),
             EarlyStopping(monitor='val_loss', patience=20),
             ModelCheckpoint(filepath='best_model.h5', monitor='val_loss', save_best_only=True)]
model.compile(loss=[tf.keras.losses.MeanSquaredLogarithmicError(),tf.keras.losses.MeanSquaredLogarithmicError()], optimizer="adam")
```

```python
history = model.fit([X_cases_train], [Y_expcases_train, Y_expfatalities_train], 
          epochs = 200, 
          batch_size = 64, 
          validation_data=([X_cases_test],  [Y_expcases_test, Y_expfatalities_test]), 
          callbacks=callbacks)
```

#### Performance of model training

### Predicting future trends

```python
#Will retrieve the number of cases and fatalities for the past 6 days from the given date
def build_inputs_for_date(country, date, df):
    start_date = date - timedelta(days=20)
    end_date = date - timedelta(days=1)
    
    str_start_date = start_date.strftime("%Y-%m-%d")
    str_end_date = end_date.strftime("%Y-%m-%d")
    df = df.query("Country_Region=='"+country+"' and Date>='"+str_start_date+"' and Date<='"+str_end_date+"'")
    
    #preparing the temporal inputs
    cases_input_data = np.transpose(np.reshape(np.asarray([df["ConfirmedCases"],
                                                 df["Fatalities"]]),
                                     (2,sequence_length)), (1,0) ).astype(np.float32)
    
    return [np.array([cases_input_data])]
```

```python
#Take a dataframe in input, will do the predictions and return the dataframe with extra rows
#containing the predictions
def predict_for_region(country, df):
    begin_prediction = "2020-04-01"
    start_date = datetime.datetime.strptime(begin_prediction,"%Y-%m-%d")
    end_prediction = "2020-05-14"
    end_date = datetime.datetime.strptime(end_prediction,"%Y-%m-%d")
    
    date_list = [start_date + timedelta(days=x) for x in range((end_date-start_date).days+1)]
    for date in date_list:
        input_data = build_inputs_for_date(country, date, df)
        result = model.predict(input_data)
        
        #just ensuring that the outputs is
        #higher than the previous counts
        result[0] = np.round(result[0])
        if result[0]<input_data[0][0][-1][0]:
            result[0]=np.array([[input_data[0][0][-1][0]]])
        
        result[1] = np.round(result[1])
        if result[1]<input_data[0][0][-1][1]:
            result[1]=np.array([[input_data[0][0][-1][1]]])
        
        #We assign the quarantine and school status
        #depending on previous values
        #e.g Once a country is locked, it will stay locked until the end
        df = df.append({"Country_Region":country, 
                        "Date":date.strftime("%Y-%m-%d"),
                        "ConfirmedCases":round(result[0][0][0]),
                        "Fatalities":round(result[1][0][0])},
                       ignore_index=True)
    return df
```

```python
#The functions that are called here need to optimise, sorry about that!
copy_df = train_data
with tqdm(total=len(list(copy_df.Country_Region.unique()))) as pbar:
    for country in copy_df.Country_Region.unique():
        copy_df = predict_for_region(country, copy_df)
        pbar.update(1)
```

```python
groundtruth_df = pd.read_csv('../datasets/train.csv')

groundtruth_df["ConfirmedCases"] = groundtruth_df["ConfirmedCases"].astype("float")
groundtruth_df["Fatalities"] = groundtruth_df["Fatalities"].astype("float")
#The country_region got modifying in the enriched dataset by @optimo, 
# so we have to apply the same change to this Dataframe.
groundtruth_df["Country_Region"] = [ row.Country_Region.replace("'","").strip(" ") if row.Province_State=="" else row.Country_Region+"_"+str(row.Province_State).replace("'","").strip(" ") for idx,row in groundtruth_df.iterrows()]

last_date = groundtruth_df.Date.unique()[-1]
```

```python
from sklearn.metrics import mean_squared_log_error
#to remove annoying warnings from pandas
pd.options.mode.chained_assignment = None

def get_RMSLE_per_region(region, groundtruth_df, display_only=False):
    groundtruth_df["ConfirmedCases"] = groundtruth_df["ConfirmedCases"].astype("float")
    groundtruth_df["Fatalities"] = groundtruth_df["Fatalities"].astype("float")
    #we only take data until the 14 of Apr 2020 as the groundtruth was not available for later dates.
    groundtruth = groundtruth_df.query("Country_Region=='"+region+"' and Date>='2020-04-01' and Date<='"+last_date+"'")
    predictions = copy_df.query("Country_Region=='"+region+"' and Date>='2020-04-01' and Date<='"+last_date+"'")
    
    RMSLE_cases = np.sqrt(mean_squared_log_error(groundtruth.ConfirmedCases.values, predictions.ConfirmedCases.values))
    RMSLE_fatalities = np.sqrt(mean_squared_log_error(groundtruth.Fatalities.values, predictions.Fatalities.values))
    
    if display_only:
        print(region)
        print("RMSLE on cases:",np.mean(RMSLE_cases))
        print("RMSLE on fatalities:",np.mean(RMSLE_fatalities))
    else:
        return RMSLE_cases, RMSLE_fatalities
    
```

```python
def get_RMSLE_for_all_regions(groundtruth_df):
    RMSLE_cases_list = []
    RMSLE_fatalities_list = []
    for region in groundtruth_df.Country_Region.unique():
        RMSLE_cases, RMSLE_fatalities = get_RMSLE_per_region(region, groundtruth_df, False)
        RMSLE_cases_list.append(RMSLE_cases)
        RMSLE_fatalities_list.append(RMSLE_fatalities)
    print("RMSLE on cases:",np.mean(RMSLE_cases_list))
    print("RMSLE on fatalities:",np.mean(RMSLE_fatalities_list))
```

```python
def display_comparison(region,groundtruth_df):
    groundtruth = groundtruth_df.query("Country_Region=='"+region+"' and Date>='2020-04-01' and Date<='2020-04-14'")
    prediction = copy_df.query("Country_Region=='"+region+"' and Date>='2020-04-01' and Date<='2020-04-14'")
    
    plt.plot(groundtruth.ConfirmedCases.values)
    plt.plot(prediction.ConfirmedCases.values)
    plt.title("Comparison between the actual data and our predictions for the number of cases")
    plt.ylabel('Number of cases')
    plt.xlabel('Date')
    plt.xticks(range(len(prediction.Date.values)),prediction.Date.values,rotation='vertical')
    plt.legend(['Groundtruth', 'Prediction'], loc='best')
    plt.show()
    
    plt.plot(groundtruth.Fatalities.values)
    plt.plot(prediction.Fatalities.values)
    plt.title("Comparison between the actual data and our predictions for the number of fatalities")
    plt.ylabel('Number of fatalities')
    plt.xlabel('Date')
    plt.xticks(range(len(prediction.Date.values)),prediction.Date.values,rotation='vertical')
    plt.legend(['Groundtruth', 'Prediction'], loc='best')
    plt.show()
```
